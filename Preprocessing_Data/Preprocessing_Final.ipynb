# MERGING DATASETS INTO SINGLE DATAFRAME

```python
all_data = olist_orders.merge(olist_items, on='order_id', how='left')
all_data = all_data.merge(olist_payments, on='order_id', how='inner')
all_data = all_data.merge(olist_reviews, on='order_id', how='inner')
all_data = all_data.merge(olist_products, on='product_id', how='inner')
all_data = all_data.merge(olist_customers, on='customer_id', how='inner')
all_data = all_data.merge(olist_sellers, on='seller_id', how='inner')
all_data = all_data.merge(olist_product_category_name,on='product_category_name',how='inner')
```

# PROCESSING OF DATE COLUMNS

```python
date_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date', 'shipping_limit_date', 'review_creation_date', 'review_answer_timestamp'] 
for col in date_columns:
    all_data[col] = pd.to_datetime(all_data[col], format='%Y-%m-%d %H:%M:%S')
```

```python
print('Duplicate data: ',all_data.duplicated().sum())
```

### Thus, we can conclude the dataframe has no duplicates 

```python
# Handling empty input in the order_approved_at column
missing_1 = all_data ['order_approved_at'] - all_data ['order_purchase_timestamp']
print (missing_1.describe ())
print ('=' * 50)
print ('Median from time of order until approved:', missing_1.median ())
```

### Median is considered because certain products were approved immediately while few were approved after a max of 60 days

```python
add_1 = all_data [all_data ['order_approved_at'].isnull()] ['order_purchase_timestamp'] + missing_1.median()
all_data ['order_approved_at'] = all_data ['order_approved_at'].fillna(add_1)
```

```python
all_data[['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']].head()
```

# Handling empty input in the order_delivered_carrier_date column

```python
missing_2 = all_data ['order_delivered_carrier_date'] - all_data ['order_approved_at']
print (missing_2.describe ())
print ('=' * 50)
print ('Median from the time apporved to submission:', missing_2.median ())

add_2 = all_data [all_data ['order_delivered_carrier_date']. isnull ()] ['order_approved_at'] + missing_2.median ()
all_data ['order_delivered_carrier_date'] = all_data ['order_delivered_carrier_date']. fillna(add_2)
```

### Median is considered because some are sent within 21 hours of the time he was approved, some are up to 107 days

# Handling empty input in the order_delivered_customer_date column

```python
missing_3 = all_data ['order_delivered_customer_date'] - all_data ['order_delivered_carrier_date']
print (missing_3.describe ())
print ('=' * 50)
print ('Median from the time it was sent to the customer received:', missing_3.median ())

add_3 = all_data [all_data ['order_delivered_customer_date']. isnull ()] ['order_delivered_carrier_date'] + missing_3.median ()
all_data ['order_delivered_customer_date'] = all_data ['order_delivered_customer_date']. fillna(add_3)
```

# Cleaning up name columns, and engineering required columns

```python

all_data['customer_city'] = all_data['customer_city'].str.title()
all_data['seller_city'] = all_data['seller_city'].str.title()
all_data['product_category_name'] = all_data['product_category_name'].str.title()
all_data['payment_type'] = all_data['payment_type'].str.replace('_', ' ').str.title()
all_data['product_category_name_english'] = all_data['product_category_name_english'].str.replace('_', ' ')
all_data['review_response_time'] = (all_data['review_answer_timestamp'] - all_data['review_creation_date']).dt.days
#Olist['delivery_against_estimated'] = (Olist['order_estimated_delivery_date'] - Olist['order_delivered_customer_date']).dt.days
#Olist['product_size_cm'] = Olist['product_length_cm'] * Olist['product_height_cm'] * Olist['product_width_cm']
all_data['order_purchase_year'] = all_data.order_purchase_timestamp.apply(lambda x: x.year)
all_data['order_purchase_month'] = all_data.order_purchase_timestamp.apply(lambda x: x.month)
all_data['order_purchase_dayofweek'] = all_data.order_purchase_timestamp.apply(lambda x: x.dayofweek)
all_data['order_purchase_hour'] = all_data.order_purchase_timestamp.apply(lambda x: x.hour)
all_data['order_purchase_day'] = all_data['order_purchase_dayofweek'].map({0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'})
all_data['order_purchase_mon'] = all_data.order_purchase_timestamp.apply(lambda x: x.month).map({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'})
```
